```python
#terms from dOmegadt

import pandas as pd
import numpy as np
import functions as f

# 1. Load the data 
file_paths = ["all_ok_scaled.csv", "freq_ok_redo_scaled.csv"]
df = pd.concat([pd.read_csv(file) for file in file_paths], ignore_index=True)

# 2. math operations
# Square root of m_companion
# We use np.sqrt() to get the square root of the entire column at once
df['sqrt_m_companion'] = np.sqrt(df['m_companion/M_Sun'])

# Add the masses 
df['total_mass'] = df['m_WR/M_Sun'] + df['m_companion/M_Sun']

# sqrt of total_mass
df['sqrt_total_mass'] = np.sqrt(df['total_mass'])

# Something to a certain power (e.g., q squared)
df['WR star Radius R_WR cubed'] = df['WR star Radius R_WR'] ** 3

# 3. Call updated function with the new column names
f.pca(
    df=df, 
    target_col='spin parameter a_spin', 
    feature_cols=[
        'm_WR/M_Sun', 
        'sqrt_m_companion',  # Use your new engineered feature
        'sqrt_total_mass',        # Use your combined feature
        'WR star Radius R_WR cubed'
    ], 
    n_components=4
)
```
